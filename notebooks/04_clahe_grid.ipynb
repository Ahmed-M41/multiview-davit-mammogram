# 04 — CLAHE Parameter Grid (Paper Fig. 3)

Visualises a 3×3 grid of CLAHE parameter combinations (clip limit × tile size) applied to a sample mammogram. This is the `RandomCLAHE` augmentation used during training.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from multiview_davit.preprocessing import read_img, to_u8

SAMPLE_IMAGE = 'data/processed/cbis_ddsm/sample.jpg'  # replace with actual path
img = to_u8(read_img(SAMPLE_IMAGE))
if img.ndim == 3:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
clip_limits = [1.0, 3.0, 5.0]
tile_sizes  = [4, 8, 12]

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
fig.suptitle('CLAHE Parameter Grid (clip limit × tile size)', fontsize=14)

for i, clip in enumerate(clip_limits):
    for j, tile in enumerate(tile_sizes):
        clahe = cv2.createCLAHE(clipLimit=clip, tileGridSize=(tile, tile))
        out = clahe.apply(img)
        axes[i][j].imshow(out, cmap='gray')
        axes[i][j].set_title(f'clip={clip}, tile={tile}×{tile}', fontsize=9)
        axes[i][j].axis('off')

plt.tight_layout()
plt.savefig('results/clahe_grid.png', dpi=150, bbox_inches='tight')
plt.show()
```
