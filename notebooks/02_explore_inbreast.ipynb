# 02 — Explore INbreast Dataset

Explores INbreast class distribution, BI-RADS categories, and ACR density distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

from multiview_davit.config import load_config
from multiview_davit.data.inbreast_labels import birads_to_binary

data_cfg = load_config('configs/data.yaml')
df = pd.read_csv(data_cfg.inbreast.csv_path)
```

```python
print(f'Total images: {len(df)}')
print(df[['Bi-Rads', 'cancer', 'ACR']].describe())
print(df['Bi-Rads'].value_counts())
```

```python
# Load codebook
codebook = pd.read_csv('codebook/inbreast_codebook.csv')
dist = pd.read_csv('codebook/inbreast_class_distribution.csv')
print(dist)
```

```python
# BI-RADS → binary label mapping
df['binary'] = df['Bi-Rads'].apply(birads_to_binary)
print('Binary distribution:')
print(df['binary'].value_counts().rename({0: 'Benign', 1: 'Malignant'}))
```
